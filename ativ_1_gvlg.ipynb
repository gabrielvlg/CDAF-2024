```python
matches_df.rename(columns={'Home Team': 'HT', 'Away Team': 'AT', 'Home Goals': 'HG', 'Away Goals': 'AG'}, inplace=True)


goal_model_data = pd.concat([matches_df[['HT', 'AT', 'HS']].assign(home=1).rename(columns={'HT':'team', 'AT':'opponent', 'HS':'goals'}),
                             matches_df[['AT', 'HT', 'AS']].assign(home=0).rename(columns={'AT':'team', 'HT':'opponent', 'AS':'goals'})])


poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=goal_model_data, family=sm.families.Poisson()).fit()


print(poisson_model.summary())
```

```python
# Set teams here
home_team='Schalke 04'
away_team='Hertha Berlin'

#Predict for Schalke 04 vs. Hertha Berlin
home_score_rate=poisson_model.predict(pd.DataFrame(data={'team': home_team, 'opponent': away_team,
                                       'home':1},index=[1]))
away_score_rate=poisson_model.predict(pd.DataFrame(data={'team': away_team, 'opponent': home_team,
                                       'home':0},index=[1]))
print(home_team + ' against ' + away_team + ' expect to score: ' + str(home_score_rate))
print(away_team + ' against ' + home_team + ' expect to score: ' + str(away_score_rate))

#Lets just get a result
home_goals=np.random.poisson(home_score_rate)
away_goals=np.random.poisson(away_score_rate)
print(home_team + ': ' + str(home_goals[0]))
print(away_team + ': '  + str(away_goals[0]))
```

```python
# Code to caluclate the goals for the match.
def simulate_match(foot_model, homeTeam, awayTeam, max_goals=10):
    home_goals_avg = foot_model.predict(pd.DataFrame(data={'team': homeTeam,
                                                           'opponent': awayTeam, 'home': 1},
                                                     index=[1])).values[0]
    away_goals_avg = foot_model.predict(pd.DataFrame(data={'team': awayTeam,
                                                           'opponent': homeTeam, 'home': 0},
                                                     index=[1])).values[0]
    team_pred = [[poisson.pmf(i, team_avg) for i in range(0, max_goals + 1)] for team_avg in
                 [home_goals_avg, away_goals_avg]]
    return (np.outer(np.array(team_pred[0]), np.array(team_pred[1])))

#Fill in the matrix
max_goals=5
score_matrix=simulate_match(poisson_model, home_team, away_team,max_goals)

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
pos=ax.imshow(score_matrix, extent=[-0.5,max_goals+0.5,-0.5,max_goals+0.5], aspect='auto',cmap=plt.cm.Reds)
fig.colorbar(pos, ax=ax)
ax.set_title('Probability of outcome')
plt.xlim((-0.5,5.5))
plt.ylim((-0.5,5.5))
plt.tight_layout()
ax.set_xlabel('Goals scored by ' + away_team)
ax.set_ylabel('Goals scored by ' + home_team)
plt.show()

#Home, draw, away probabilities
homewin=np.sum(np.tril(score_matrix, -1))
draw=np.sum(np.diag(score_matrix))
awaywin=np.sum(np.triu(score_matrix, 1))
```

## Questão 5
- Utilize o modelo treinado para simular os placares de todos os jogos da temporada.
- Construa uma tabela de classificação em cima dos resultados esperados. Considere que jogos com uma diferença esperada de gols < 0.5 são empates.
- Compare a tabela real com a simulada. Onde estão as principais diferenças entre elas? E similaridades? Como você interpreta essa comparação? É possível identificar aspectos que o modelo subestima/superestima? É possível identificar tiveram campanhas "sortudas" ou "azaradas"?

```python
def simular_placar(poisson_model, home_team, away_team):
    home_goals_avg = poisson_model.predict(pd.DataFrame(data={'team': home_team,
                                                               'opponent': away_team,'home':1},
                                                         index=[1])).values[0]
    away_goals_avg = poisson_model.predict(pd.DataFrame(data={'team': away_team,
                                                               'opponent': home_team,'home':0},
                                                         index=[1])).values[0]
    team_goals_sim = np.random.poisson(home_goals_avg)
    opponent_goals_sim = np.random.poisson(away_goals_avg)
    return team_goals_sim, opponent_goals_sim


placares_simulados = []

for index, row in matches_df.iterrows():
    home_team = row['HT']
    away_team = row['AT']
    home_goals, away_goals = simular_placar(poisson_model, home_team, away_team)
    placares_simulados.append({'HT': home_team, 'AT': away_team,
                               'HS': home_goals, 'AS': away_goals})


placares_simulados_df = pd.DataFrame(placares_simulados)


print(placares_simulados_df)
```

```python

def calcular_pontos(home_goals, away_goals):
    if home_goals > away_goals:
        return 3  # Vitória
    elif home_goals < away_goals:
        return 0  # Derrota
    else:
        return 1  # Empate


placares_simulados_df['Pontos_HomeTeam'] = placares_simulados_df.apply(lambda row: calcular_pontos(row['HS'], row['AS']), axis=1)
placares_simulados_df['Pontos_AwayTeam'] = placares_simulados_df.apply(lambda row: calcular_pontos(row['AS'], row['HS']), axis=1)


tabela_classificacao = {}

for index, row in placares_simulados_df.iterrows():
    home_team = row['HT']
    away_team = row['AT']
    home_points = row['Pontos_HomeTeam']
    away_points = row['Pontos_AwayTeam']

    if home_team not in tabela_classificacao:
        tabela_classificacao[home_team] = {'Pontos': 0, 'Posicao': 0}
    if away_team not in tabela_classificacao:
        tabela_classificacao[away_team] = {'Pontos': 0, 'Posicao': 0}

    tabela_classificacao[home_team]['Pontos'] += home_points
    tabela_classificacao[away_team]['Pontos'] += away_points


classificacao = pd.DataFrame.from_dict(tabela_classificacao, orient='index')
classificacao['Posicao'] = classificacao['Pontos'].rank(method='min', ascending=False).astype(int)


print(classificacao.sort_values(by='Posicao'))
```
